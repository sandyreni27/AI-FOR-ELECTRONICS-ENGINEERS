```python
import numpy as np
import matplotlib.pyplot as plt

Fs=100
t=np.linspace(0,1,Fs)

#Message and carrier signals
fm=5
fc=50

message=np.cos(2*np.pi*fm*t)
carrier=np.cos(2*np.pi*fc*t)
#1.Amplitude modulation
mu=0.7
am_signal=(1+mu*message)*carrier

#2.Frequency modulation
kf=20
fm_signal=np.cos(2*np.pi*fc*t+kf*np.sin(2*np.pi*fm*t))

#3.Phase modulation
kp=1.0
pm_signal=np.cos(2*np.pi*fc*t+kp*message)

plt.figure(figsize=(12,10))

plt.subplot(4,1,1)
plt.plot(t,message)
plt.title("message Signal (5hz)")
plt.grid(True)

plt.subplot(4,1,2)
plt.plot(t,am_signal)
plt.title("AM Signal(mu=0.7)")
plt.grid(True)

plt.subplot(4,1,3)
plt.plot(t,fm_signal)
plt.title("FM Signal(Frequency Modulation)")
plt.grid(True)

plt.subplot(4,1,4)
plt.plot(t,pm_signal)
plt.title("PM Signal(Phase Modulation)")
plt.grid(True)

plt.tight_layout()
plt.show()
```
