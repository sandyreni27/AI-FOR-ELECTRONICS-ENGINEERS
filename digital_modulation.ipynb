```python
import numpy as np
import matplotlib.pyplot as plt

Fs=1000         #sampling frequency
Tb=0.1          #bit duration
t=np.linspace(0,Tb,int(Fs*Tb))         #time for 1 bit

fc=20           #carrier frequency
f1=20           #FSK '1' frequency
f0=5            #FSK '0' frequency
#----------------------------------------------
#Generate Random Bits
#---------------------------------------------
bits=np.random.randint(0,2,8)   #8 random bits
print("Bits:",bits)
#---------------------------------------------
#1.ASK Modulation
#--------------------------------------------
ask_signal=[]
for b in bits:
    if b==1:
        ask_signal.extend(np.cos(2*np.pi*fc*t))   #high amplitude
    else:
        ask_signal.extend(0.3*np.cos(2*np.pi*fc*t))      #low amplitude

ask_signal=np.array(ask_signal)
#----------------------------------------------
#3.BPSK Modulation
#----------------------------------------------
bpsk_signal=[]
for b in bits:
    if b==1:
        bpsk_signal.extend(np.cos(2*np.pi*fc*t))
    else:
        bpsk_signal.extend(-np.cos(2*np.pi*fc*t))

bpsk_signal=np.array(bpsk_signal)
#------------------------------------------------
#Time axis for whole signal
#------------------------------------------------
T=np.linspace(0,Tb*len(bits),len(ask_signal))
#------------------------------------------------
# #plot All signals
#------------------------------------------------
plt.figure(figsize=(12,10))

plt.subplot(4,1,1)
plt.step(range(len(bits)),bits,where='mid')
plt.title("Digital Bits")
plt.ylim(-0.5,1.5)
plt.grid(True)

plt.subplot(4,1,2)
plt.plot(T,ask_signal)
plt.title("ASK Modulation")
plt.grid(True)

plt.subplot(4,1,4)
plt.plot(T,bpsk_signal)
plt.title("BPSK Modulation")
plt.grid(True)

plt.tight_layout()
plt.show()
```
